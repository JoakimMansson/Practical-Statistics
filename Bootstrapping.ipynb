```python
import numpy as np
import matplotlib.pyplot as plt
```

## Normal Distributed Data

```python
num_data_points = 1000
mu = 0
sigma = 2
data = np.random.normal(loc=mu, scale=sigma, size=num_data_points)
plt.hist(data)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

## Confidence Interval

```python
n = num_data_points//10
num_samples = 10
bootstrap_means = []
for i in range(num_samples):
    sample_mean = np.mean(np.random.choice(data, n, replace=True))
    bootstrap_means.append(sample_mean)

print(f"Shape bootstrap_mean: {np.array(bootstrap_means).shape}")
standard_error = np.std(bootstrap_means, axis=0)
print(f"Standard error for bootstrap: {standard_error}")
```

```python
lower_bound = np.percentile(bootstrap_means, 2.5)
upper_bound = np.percentile(bootstrap_means, 97.5)
print(f"95% confident that the population mean is between {lower_bound} <-> {upper_bound}")
```
